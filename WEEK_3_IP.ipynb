```python
# Import the relevant libraries
import pandas as pd 
import numpy as np
```

```python
# Import the Telecom_dataset csv file rename it to day1 and view the DataFrame
day1=pd.read_csv('/content/Telcom_dataset.csv')
day1.head()
```

```python
# Import the Telecom_dataset2 csv file, rename it to day2 and view the DataFrame
day2=pd.read_csv('/content/Telcom_dataset2.csv')
day2.head()
```

```python
# Import the Telecom_dataset2 csv file, rename it to day3 and view the DataFrame
day3=pd.read_csv('/content/Telcom_dataset3.csv')
day3.head()
```

```python
# Import data from cell_geo url link 
url = 'https://drive.google.com/file/d/1ABZux280OjL3yWcOn8BDA_f5QsyO0QPU/view'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
cells_geo = pd.read_csv(path, header=0,index_col=0, sep=';') 
cells_geo.head()
```

```python
# Rename column names of day1, day2 and day3 
# In day1 column we will change the names PRODUTC	to PRODUCT and DATETIME to DATE_TIME
# In day2 column we will change the names DW_A_NUMBER to DW_A_NUMBER_INT and DW_B_NUMBER to DW_B_NUMBER_INT
# In day3 column we will change the names CELLID to CELL_ID and SIET_ID to SITE_ID

day1.rename(columns={'PRODUTC': 'PRODUCT','DATETIME':'DATE_TIME'}, inplace=True)
day2.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER' : 'DW_B_NUMBER_INT'}, inplace=True )
day3.rename(columns={'CELLID': 'CELL_ID','SIET_ID': 'SITE_ID'}, inplace=True)
```

```python
# Confirm changes in day1
day1.head(2)
```

```python
# Confirm changes in day2
day2.head(2)
```

```python
# Confirm changes in day3
day3.head(2)
```

```python
# We will merge the 3 DataFrames day1,day2,day3 into days
days=pd.concat([day1,day2,day3], axis=0,ignore_index=True)
days.head()
```

```python
# Merge the days and cells_geo DataFrames and save as merged
merged=pd.merge(days,cells_geo, left_on='SITE_ID', right_on='SITE_CODE')
merged
```

```python
# Check for null values in merged
merged.info()
```

```python
# Delete any row containing a null value 
cleaned_merged=merged.dropna(axis=1,how='any')
cleaned_merged.head(4)
```

```python
# Confirm the changes made above 
cleaned_merged.info()
```

```python
# Determine the cities with the highest customer activity
cleaned_merged.groupby(['VILLES','REGION']).size().sort_values(ascending=False)
```

```python
# Determine the cities spending the most money on MTN services
cleaned_merged.groupby(['VILLES','REGION'])['VALUE'].sum().sort_values(ascending=False)
```
